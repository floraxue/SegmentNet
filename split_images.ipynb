```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_ubyte
import matplotlib
%matplotlib inline
import os
from os.path import join
```

```python
image_path = '../data/dataset_1/raw_image_100/{:04d}.tif'.format(1)

print(image_path)
```

```python
x_test = imread(image_path)
x_test = img_as_ubyte(x_test)
imshow(x_test)
x_test.shape
```

```python
target_size = 256
stride = 64
num_crops = (x_test.shape[0] - target_size) / stride + 1   # num_crops per axis. total_crops = num_crops**2
print(num_crops)
num_crops = int(num_crops)
```

```python
# [i_start, i_end)
save_dir = '../data/dataset_1/raw_crop_image_100/{:04d}/'.format(1)
os.makedirs(save_dir, exist_ok=True)
for row in range(num_crops):
    for col in range(num_crops):
        jstart, jend = col * stride, col * stride + target_size
        istart, iend = row * stride, row * stride + target_size
        curr_crop = x_test[istart:iend, jstart:jend].copy()
        imsave(join(save_dir, 'row_{}_col_{}_crop.tif'.format(row, col)), curr_crop)
```

```python
##################### Formal code below #####################
# Configurations
input_img_dir = '../data/dataset_1/label_RGB_100/'
all_save_dir = '../data/dataset_1/label_RGB_crop_100/'
os.makedirs(all_save_dir, exist_ok=True)
input_shape = 2560
target_size = 256
stride = 64
num_crops = (input_shape - target_size) / stride + 1   # num_crops per axis. total_crops = num_crops**2
print("Num_crops", num_crops)
num_crops = int(num_crops)

import warnings
# with warnings.catch_warnings():
#     warnings.simplefilter("ignore")
#     img_as_ubyte(image)
    
for i in range(1, 101):
    
    image_path = join(input_img_dir, '{:04d}.tif'.format(i))
    print("Now saving " + image_path)
    img = imread(image_path)
    img = img_as_ubyte(img)
    
    for row in range(num_crops):
        for col in range(num_crops):
            jstart, jend = col * stride, col * stride + target_size
            istart, iend = row * stride, row * stride + target_size
            curr_crop = img[istart:iend, jstart:jend].copy()
            save_path = join(all_save_dir, '{:04d}_row_{}_col_{}_crop_rgb_mask.tif'.format(i, row, col))
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                imsave(save_path, curr_crop)
```
